```python
import numpy as np
import pymc3 as pm
import arviz as az

observed = np.random.binomial(500, p=0.4, size=30)

with pm.Model() as model:
    mu = pm.DiscreteUniform('mu', lower=100, upper=1000)
    p = pm.Beta('p', alpha=1, beta=1)
    N = pm.Binomial('N', n=mu, p=p, observed=observed)

    trace = pm.sample(trace=[mu, p])

az.plot_trace(trace)
```
